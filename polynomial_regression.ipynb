```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = 8*np.random.rand(200,1)-4
y1 = 0.8*X1**2+0.9*X1+2+np.random.randn(200,1)
```

```python
plt.scatter(X1,y1)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X1,y1,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_preds = lr.predict(X_valid)
r2_score(y_valid,y_preds)
```

```python
plt.plot(X_train,lr.predict(X_train),c='r')
plt.scatter(X1,y1)
plt.show();
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly.fit_transform(X_train)
X_valid_poly = poly.transform(X_valid)
```

```python
lr.fit(X_train_poly,y_train) #never apply polynomial regression in 'y' target column
```

```python
ypreds = lr.predict(X_valid_poly)
r2_score(y_valid,ypreds)
```

```python
print(X_train[0]) #shows the normal first data point
print(X_train_poly[0]) #shows the converted data point into a polynomial point with X^1, X^2
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new = np.linspace(-4,4,200).reshape(200,1)
X_new_poly = poly.transform(X_new)
```

```python
plt.plot(X_new, lr.predict(X_new_poly) , c='r',label='Polynomial Regression')
plt.scatter(X_train,y_train,c='b',label='Training Data')
plt.scatter(X_valid,y_valid,label='Validation Data')
plt.legend()
plt.show();
```

```python
print("Input:",poly.n_features_in_)
print("Output:",poly.n_output_features_)
print("Power:",poly.powers_)
```
