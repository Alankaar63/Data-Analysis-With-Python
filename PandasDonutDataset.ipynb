```python
n = 2 * np.pi * np.random.rand(10000)
```

```python
# x1 = 10 * np.cos(n)
# x2 = 2 * np.sin(n)
```

```python
x1_outer = 5*np.cos(n)
x2_outer = 5*np.sin(n)

x1_inner = 3*np.cos(n)
x2_inner = 3*np.sin(n)

x1 = np.concatenate((x1_outer,x1_inner))
x2 = np.concatenate((x2_outer,x2_inner))
```

```python
df = pd.DataFrame(x1, columns=['x1'])
df['x2'] = x2
df.head()
```

```python
df['x1^2'] = df['x1'] ** 2
df['x2^2'] = df['x2'] ** 2
df['x1 x x2'] = df['x1'] * df['x2']
df.head()
```

```python
x_inner = np.zeros(10000)
y_outer = np.ones(10000)
df['y'] = np.concatenate((x_inner, y_outer))
df.head()
```

```python
final_data = df[['x1', 'x2', 'x1^2', 'x2^2', 'x1 x x2', 'y']]
final_data.to_csv('donut_dataset.csv', index=False, header=None)
```
