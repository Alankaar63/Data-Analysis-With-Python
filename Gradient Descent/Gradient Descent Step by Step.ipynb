```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression

X,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1,noise= 80, random_state = 13)
```

```python
plt.scatter(X,y);
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

```python
lr.predict(X)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red')
plt.show();
```

```python
lr.intercept_ #b
```

```python
lr.coef_ #m
```

```python
#now lets appply gradient descend by taking m as constant -57.99
# and lets take any random value of b = 2

y_pred = ((78.35*X)+2).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color='green',label=2)
plt.legend();
plt.show();
```

```python
slope = -2*np.sum(y-(lr.coef_)*X.ravel()-2)
slope
```

```python
step_rate = 0.1*slope
step_rate
```

```python
bnew = 2 - step_rate
bnew
```

```python
y_pred1 = ((78.35*X)+ bnew).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color='green',label=2)
plt.plot(X,y_pred1,color='blue',label=bnew)
plt.legend();
plt.show();
```

```python
loss_slope = -2*np.sum(y-78.35*X.ravel()-bnew)
loss_slope
```

```python
n = 0.1
nslope = n*loss_slope
bnew = bnew-nslope

bnew
```

```python
y_pred2 = ((78.35*X) + bnew).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='#00a65a',label= 'b1')
plt.plot(X,y_pred1,color='#A3E4D7',label='b2')
plt.plot(X,y_pred2,color='#A3E4D7',label='b3')
plt.legend();
plt.show();
```

```python
m = 78.35
b = -100
n = 0.01

epochs = 10
for i in range(epochs):
  loss_slope = -2*np.sum(y-m*X.ravel()-b)
  nslope = n*loss_slope
  b = b-nslope
  print(b)
  plt.plot(X,(((78.35*X)+b).reshape(4)),color='#00a65a')

plt.scatter(X,y)
```
