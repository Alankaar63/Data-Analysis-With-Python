```python
m = 27.39
b = -0.82
```

```python
class myGDRegressor:
  def __init__(self,learning_rate,epochs):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.m = m
    self.b = b

  def fit(self,X,y):
    for i in range(self.epochs):
      loss_slope = -2*np.sum(y-self.m*X.ravel()-self.b)
      nslope = self.learning_rate*loss_slope
      self.b = self.b - nslope

    print(self.b)

```

```python
gd = myGDRegressor(0.01,100)
```

```python
gd.fit(X,y)
```
