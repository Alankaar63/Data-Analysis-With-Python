```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from scipy.signal import convolve2d
from scipy.stats import norm
```

```python
!wget https://i.pinimg.com/564x/28/14/d3/2814d3f4b243a87d2150363729f3d732.jpg
```

```python
im = Image.open('2814d3f4b243a87d2150363729f3d732.jpg')
```

```python
gray = np.mean(im,axis=2)
```

```python
grayscale_image = plt.imshow(gray,cmap='gray');
```

```python
gray_image = np.array(grayscale_image)
```

```python
Hx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
Hy= np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
Gx = convolve2d(gray,Hx,mode='same')
Gy = convolve2d(gray,Hy,mode='same')
```

```python
G = np.sqrt(Gx**2,Gy**2)
```

```python
plt.subplot(1,2,1)
plt.imshow(gray,cmap='gray')
plt.title('Normal Image')
plt.subplot(1,2,2)
plt.imshow(G,cmap='gray')
plt.title('Edge Detection')
plt.show();
```
