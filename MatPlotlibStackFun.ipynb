#matplotlib Exercise

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X = np.random.uniform(-1,1,(500,2))
```

```python
y = np.where(np.sign(X[:,0]) == np.sign(X[:,1]),0,1)
```

```python
plt.scatter(X[y==0][:,0],X[y==0][:,1],color='yellow',edgecolor='y')
plt.scatter(X[y==1][:,0],X[y==1][:,1],color='blue',edgecolor = 'b')
plt.title('XOR dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.legend()
plt.show();
```

#matplotlib Exercise 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X = np.random.uniform(-1,1,(1000,2))
```

```python
Y = np.where(np.sign(X[:,0])== np.sign(X[:,1]),1,0) #class o -> 1 , class 1-> 0
```

```python
plt.scatter(X[Y==0][:,0], X[Y==0][:,1],color='cyan',edgecolor='c')
plt.scatter(X[Y==1][:,0],X[Y==1][:,1],color='red',edgecolor='r')
plt.title('XNOR dataset')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.legend()
plt.show();
```
