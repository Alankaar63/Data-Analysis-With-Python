```python
df = pd.read_csv('Ecommerce_Sales_Prediction_Dataset.csv')
df.head()
```

```python
#filling in the missing values
print(df.isnull().sum())
```

```python
df.drop('Date',axis=1)
```

```python
df['sales'] = df['Units_Sold'].apply(lambda x: 'High Sales' if x>=25 else 'Low Sales' )
```

```python
df.head()
```

```python
cleaned_data = df.drop('Date',axis=1)
cleaned_data.replace({'High Sales':1, 'Low Sales':0},inplace=True)
```

```python
#splitting the dataset
X = cleaned_data.drop('sales',axis=1)
y = cleaned_data['sales']
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.25,random_state=0)
print(X_train.shape)
print(X_valid.shape)
print(y_train.shape)
print(y_valid.shape)
```

```python
num_cols = X_train.select_dtypes(exclude=['object']).columns
cat_cols = X_train.select_dtypes(include=['object']).columns
print(num_cols)
print(cat_cols)
```

```python
#making the pipeline

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer

#imputing numerical data
numerical_transformer = SimpleImputer(strategy='constant')

categorical_transformer = Pipeline(steps=[('imputer',SimpleImputer(strategy='most_frequent')),('onehot',OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(transformers=[('num',numerical_transformer,num_cols),
                                 ('cat',categorical_transformer,cat_cols)])
```

#**Using RandomForestClassifier**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=150,random_state=0)
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score,confusion_matrix
from sklearn.model_selection import cross_val_score

my_pipeline = Pipeline(steps=[('preprocessor',preprocessor),('model',model)])
cv_score = cross_val_score(my_pipeline,X_train,y_train,cv=10,scoring='accuracy')
print(cv_score)
print(cv_score.mean())
my_pipeline.fit(X_train,y_train)
preds = my_pipeline.predict(X_valid)
score = mean_absolute_error(y_valid,preds)
print("MAE:",score)
print("Accuracy:",accuracy_score(y_valid,preds))
print("Confusion Matrix:\n",confusion_matrix(y_valid,preds))
```
